# What is the Bayes Rule?

A way for us to find the probability of event given data



## The essence

## Perfect inference engine

Bayesian inference does not guarantee the correct answer. It provides probability of each possible answer to be true. One can use that result to find the most probable answer. It gives us an informed guess.

What this means is that the Bayesian inference engine can be wrong, but it is provably less wrong than any other i.e. it is the best one can do.

## Frequentist point of view

I use my remote to unlock the car and listen for a sound. Sure enough, I hear a sound from not far away. I have a mental model of the parking area, so upon hearing the sound I infer the area I must search to find my car.

## Bayesian point of view

I use my remote to unlock the car and listen for a sound. Sure enough, I hear a sound from not far away. I have a mental model of the parking area **and a memory of where I parked previously (my prior)**. I combine those two inferences to identify the area I must search to find my car.


| Frequentist                                              | Bayesian                                               |
|----------------------------------------------------------|--------------------------------------------------------|
| Data are repeatable random sample - there is a frequency | Data are observed from realised sample                 |
| Underlying parameters remain constant during the process | Parameters are unknown and described probabilistically |
| **Parameters are fixed**                                 | **Data is fixed**                                      |

# Back to the matter at hand


Ivan's face was pretty grim. It was pretty obvious to him that his engine was making loud noises and that wasn't a good thing! Bad timing, he thought, just when I needed a load of cash to sponsor my social media marketing campaign.

The mechanic's face was telling a completely different story, though! Maybe he was happy that a lot of cash will end up in his pocket (BMW repairs were pretty expensive - everybody knew that one). He was considering only two options - completely worn out brakes and bearing failure.

Ivan asked with a trembling voice: "Zdravko, what is it and how much to repair? Save me, brother!". Zdravko smiled and said nothing. His wrench was close by, so he took it.

What was happening in Zdravko's head you might ask?

95% of the engines that has bearing failure have the same loud noise as that of Ivan's car. So the probability of hearing knocking noises from your engine given that your car has bearing failure is 95%. 

But you don't want to know what is the probability of hearing those sounds! **You want to know what is the probability of having bearing failure!**

Zdravko knew that 70% of the cars with completely work out breaks make the same exact noise. Zdravko combined the possible causes implied by the noise with his prior experience and got the following:

This mechanic has been working with cars for more than 15 years and has a vast experience of similar cases. Using his knowlege he estimated that the probability of having the noise given that the car has bearing failure is 95%. Which can be written as:

$$\tag{1} p(\text{engine noise}\,|\,\text{bearing failure})=0.95$$

Similarly, he estimated the probability of having the noise given that the brakes are worn out as:

$$\tag{2} p(\text{engine noise}\,|\,\text{worn out brakes})=0.7$$

These equations show us the **likelihood** of the car having bearing failure or worn out brakes, respectively.

Now was the time to use Zdravko's vast experience. He estimated that every 10 out of random 1000 cars (or around 1%) have a bearing failure. Which can be expressed as:

$$\tag{3} p(\text{bearing failure})=0.01$$

This represents Zdravko's prior knowledge (or prior probability) about bearing failure before even seeing the car itself.

Zdravko is no fool! He knew that he has to combine some of the numbers somehow. One last thing: he estimated that every 80 out of 1000 (or 0.08) cars have engine noises.

And now the tricky part (this might require a leap of faith) - Zdravko knew about the Bayes Rule! Furthermore, he knew how to apply it:

$$\tag{4} p(\text{bearing failure}\,|\,\text{engine noise}) = \frac{p(\text{engine noise}\,|\,\text{bearing failure}) \, p(\text{bearing failure})}{p(\text{random car has engine noise})}$$

Zdravko calculated the probability of having a bearing failure:

$$\tag{5} p(\text{bearing failure}\,|\,\text{engine noise}) = (0.95 \times 0.01)\,/\,0.08 \approx 0.11875 $$

or below 12% - this is called **posterior probability**. 

Now 12% sounds like unlikely event but Zdravko had to compare it against the other option - worn out brakes.

He estimated that every 100 out of random 1000 cars (or around 10%) have worn out brakes. Which can be written as:

$$\tag{6} p(\text{worn out brakes})=0.1$$

So the probability of having worn out brakes given the engine noise is:

$$\tag{5} p(\text{worn out brakes}\,|\,\text{engine noise}) = (0.7 \times 0.1)\,/\,0.08 \approx 0.875 $$

or roughly 88% (Forgive Zdravko his rounding errors, he is doing this in his head). Note that the probabilities sum up almost to 1 (Coincidence? See below), so Zdravko was doing good.

Now for some Bayesian model comparison Zdravko tried this:

$$\frac{p(\text{worn out brakes}\,|\,\text{engine noise})}{p(\text{bearing failure}\,|\,\text{engine noise})} \approx \frac{0.875}{0.11875} \approx 7.368$$

So it is 7+ times more likely for the car to have worn out brakes!

```python
# Calculate likelihood x prior

p1 = (0.95 * 0.01)
p2 = (0.7 * 0.1)
print(p1, p2)

# Find the normalizing constant (0.08 in the example)

norm = 1 / sum([p1, p2])
p1_norm = p1 * norm
p2_norm = p2 * norm

print(p1 / p1_norm)
print(p2 / p2_norm)

norm_const = p1 / p1_norm

# Find the posterior probabilities (likelihood x prior) / norm constant

print(p1 / norm_const)
print(p2 / norm_const)
```

# Using some Python magic

Let's setup our environment

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import scipy.stats as stats
from pylab import rcParams
sns.set(style='ticks', palette='Set2')
%matplotlib inline
rcParams['figure.figsize'] = 12, 8
rcParams['legend.fontsize'] =  20
```

## Worn out brakes

Let's define a grid of 100 values between 0 and 1:

```python
x_range = np.linspace(0, 1, 100)
```

Zdravko found that the actual probability of having worn out brakes is 87.5%. We will draw 50 samples from a binomial distribution with $p=0.875$. This will represent our likelihood.

```python
likelihood_sample = np.random.binomial(1, .875, size=50)
likelihood = np.array([np.product(stats.bernoulli.pmf(likelihood_sample, r)) for r in x_range])
```

Zdravko estimated that the prior probability for a random car to have worn out brakes is 10%.  We will draw 1000 samples from this distribution and ask what is the probability of observing every value in our range.

```python
prior_sample = np.random.binomial(1, .1, size=1000)
prior = np.array([np.product(stats.bernoulli.pmf(prior_sample, r)) for r in x_range])
# Make our prior sum to 1
prior = prior / np.sum(prior)
```

Finally, let's find the posterior distribution of having worn out brakes

```python
# Posterior = Prior x Likelihood 
post = np.multiply(prior, likelihood)
# Make our posterior sum to 1
post = post / np.sum(post)
```

Let's see what we got:

```python
fig, axes = plt.subplots(3, 1, sharex=True)
axes[0].plot(x_range, likelihood)
axes[0].set_title("Sampling Distribution")
axes[1].plot(x_range, prior)
axes[1].set_title("Prior Distribution")
axes[2].plot(x_range, post)
axes[2].set_title("Posterior Distribution")
sns.despine()
plt.tight_layout()
```

Sweet! Our posterior hasn't moved away much from the prior, even though our prior is centered approximately around 0.9. What happens if we take 500 samples from our sampling distribution? (Will have to wrap our code in a function)

```python
def find_posterior(true_p=.875, sample_n=50, prior_p=.1, prior_n=1000):
    x_range = np.linspace(0, 1, 100)
    likelihood_sample = np.random.binomial(1, true_p, size=sample_n)
    likelihood = np.array([np.product(stats.bernoulli.pmf(likelihood_sample, r)) for r in x_range])

    prior_sample = np.random.binomial(1, prior_p, size=prior_n)
    prior = np.array([np.product(stats.bernoulli.pmf(prior_sample, r)) for r in x_range])
    # Make our prior sum to 1
    prior = prior / np.sum(prior)

    # Posterior = Prior x Likelihood 
    post = np.multiply(prior, likelihood)
    # Make our posterior sum to 1
    post = post / np.sum(post)
    return (x_range, prior, likelihood, post)

def plot_bayes_rule(true_p=.875, sample_n=50, prior_p=.1, prior_n=1000):
    x_range, prior, likelihood, post = find_posterior(true_p, sample_n, prior_p, prior_n)
    fig, axes = plt.subplots(3, 1, sharex=True)
    axes[0].plot(x_range, likelihood)
    axes[0].set_title("Sampling Distribution")
    axes[1].plot(x_range, prior)
    axes[1].set_title("Prior Distribution")
    axes[2].plot(x_range, post)
    axes[2].set_title("Posterior Distribution")
    sns.despine()
    plt.tight_layout()
```

```python
plot_bayes_rule(sample_n=500)
```

Our posterior have moved quite a bit to the right (away from the prior) since the sample size has increased tenfold (from 50 to 500) and our sampling distribution is centered around high probability values, thus the evidence provided by the likelihood is much stronger now.

## Bearing failure

Zdravko found that the actual probability of having bearing failure is 11.9%. We will draw 50 samples from a binomial distribution with $p=0.119$. He also estimated that prior probability for a random car to have bearing failure is 1%.

```python
plot_bayes_rule(true_p=0.119, sample_n=50, prior_p=0.01)
```

The posterior almost hasn't moved. What if we increase the sample to 500?

```python
plot_bayes_rule(true_p=0.119, sample_n=500, prior_p=0.01)
```

Our posterior hasn't moved much to the right (away from the prior) even though the sample size increased (again tenfold). That looks ok, our sampling distribution and prior are centered around low probability values.

## Comparing posterior distributions

Ok, we have our results. Now let's compare them (Will use the distributions with 500 samples):

```python
x_range, _, _, brakes_post = find_posterior(sample_n=500)
_, _, _, bearing_post = find_posterior(true_p=0.119, sample_n=500, prior_p=0.01)
```

```python
f1 = plt.plot(x_range, brakes_post)
f2 = plt.plot(x_range, bearing_post)
plt.legend(["Worn out brakes", "Bearing failure"])
sns.despine()
plt.tight_layout();
```
